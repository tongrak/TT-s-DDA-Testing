```python
def containAll(input:str, criteas:list)->bool:
    res = 0
    for cr in criteas:
        if cr in input: res+=1
    return res == len(criteas)
```

```python
from functools import reduce 

class pattern:
    def __init__(self) -> None:
        self.seqList = [0,0,0,0,0]

    def canIncrease(self)->bool:
        if self.seqList[4] > 2: return False
        else: return True

    def getPatt(self)->list:
        return self.seqList.copy()

    def Increase(self)->None:
        if self.canIncrease():
            self.seqList[0] += 1;
            for i in range(4):
                if self.seqList[i] == 3: 
                    self.seqList[i] = 0
                    self.seqList[i+1] += 1

gametypeTexts = ["gametype1","gametype2","gametype3"]
minTexts = ["min_<--0.5", "min_-0.5<->0", "min_0->"]
meanTexts = ["mean_<--0.5", "mean_-0.5<->0.5", "mean_0.5->"]
maxTexts = ["max_<-0", "max_0<->0.5", "max_0.5->"]
stdTexts = ["std_<-0.25", "std_0.25<->1", "std_1->"]
moodList = ["stressed", "fine", "bored"]

inputFile = open("../placeholderData/tier2.data",'r')
inTexts = inputFile.readlines();
outputFile = open("../placeholderData/tier2Interpret.data", 'w')

thePatt = pattern();
while(thePatt.canIncrease()):
    currPatts = thePatt.getPatt();
    componList = [gametypeTexts, minTexts, meanTexts, maxTexts, stdTexts]
    currCris = list(map(lambda x,y: x[y], componList, currPatts));
    # currToFind = reduce(lambda x,y: str(x)+"    "+str(y), currCris)
    
    moodNums = [0,0,0]
    moodCris = [currCris.copy(),currCris.copy(),currCris.copy()]
    for i in range(3):
        moodCris[i].append(moodList[i])

    for inStr in inTexts:
        for i in range(3):
            if containAll(inStr, moodCris[i]): moodNums[i] += 1

    sumVal = sum(moodNums)
    probStrs = []
    if sumVal == 0:
        probStrs = ['0','0','0']
    else:
        probFunc = lambda x: str(round(x/sumVal,2))
        probStrs = list(map(probFunc, moodNums))
        moodStrs = []
        for iT in moodCris:
            moodStrs.append(reduce(lambda x,y: str(x)+"|"+str(y), iT))
        chosenIndex = probStrs.index(max(probStrs))
        outputFile.write("\n" + moodStrs[chosenIndex])

    thePatt.Increase()


inputFile.close()
outputFile.close()
```
